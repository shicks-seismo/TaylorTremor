```python
from obspy import read_inventory
import pandas as pd
from obspy.clients.filesystem.sds import Client
from obspy import UTCDateTime
import numpy as np
import matplotlib.pyplot as plt
from obspy.signal.filter import envelope
import matplotlib.mlab as mlab
from obspy.signal.util import next_pow_2
from datetime import timedelta, datetime
import matplotlib.dates as mdates
from obspy import Stream
from numpy import trapz
from scipy.integrate import simpson
from matplotlib.dates import DateFormatter
import pandas as pd
import matplotlib.font_manager as font_manager
import matplotlib
font_dir = ['/Users/sph1r17/Downloads/Source_Sans_Pro']
for font in font_manager.findSystemFonts(font_dir):
    font_manager.fontManager.addfont(font)
matplotlib.rc('font', family='Source Sans Pro') 
```

# 1. Get data for each concert data

```python
c = Client("/Volumes/WembleyData/")

ot_fri = UTCDateTime("2024-06-21T18:24:00.000Z")
ot_sat = UTCDateTime("2024-06-22T18:20:00.000Z")
ot_sun = UTCDateTime("2024-06-23T17:29:00.000Z")

st_sun = c.get_waveforms(network="TS", station="WEM05", location="00", channel="CH*", starttime=ot_sun-(1.5*60*60), endtime=ot_sun+(4.5*60*60))
st_sat = c.get_waveforms(network="TS", station="WEM05", location="00", channel="CH*", starttime=ot_sat-(1.5*60*60), endtime=ot_sat+(4.5*60*60))
st_fri = c.get_waveforms(network="TS", station="WEM05", location="00", channel="CH*", starttime=ot_fri-(1.5*60*60), endtime=ot_fri+(4.5*60*60))

# Decimate to 50 sps since we are really only interested in <=20 Hz
st_sun.decimate(10)
st_sat.decimate(10)
st_fri.decimate(10)

# Get instrument responses
inv = read_inventory("/Users/sph1r17/Downloads/TS.xml", format="STATIONXML")
```

# 2. Restitute the data to ground displacement and do some pre-filtering etc.

```python
st_sun_ = st_sun.copy()
st_sat_ = st_sat.copy()
st_fri_ = st_fri.copy()

st_sun_.remove_response(inventory=inv, output="DISP")
st_sat_.remove_response(inventory=inv, output="DISP")
st_fri_.remove_response(inventory=inv, output="DISP")

st_sun_.detrend(type="demean")
st_sat_.detrend(type="demean")
st_fri_.detrend(type="demean")
st_sun_.detrend(type="linear")
st_sat_.detrend(type="linear")
st_fri_.detrend(type="linear")

st_sun_.filter("bandpass", freqmin=1, freqmax=10)
st_sat_.filter("bandpass", freqmin=1, freqmax=10)
st_fri_.filter("bandpass", freqmin=1, freqmax=10)

st_sun_.taper(max_percentage=0.03)
st_sat_.taper(max_percentage=0.03)
st_fri_.taper(max_percentage=0.03)
```

# 3. Plot daily comparisons for Z-component 

```python
fig = plt.figure(figsize=(8, 6))
ax = plt.subplot()
ax.plot(st_sun_[2].times(reftime=ot_sun)/60, 1000*st_sun_[2].data+0.015, alpha=0.8, label="Sunday", lw=0.4)
ax.plot(st_sat_[2].times(reftime=ot_sat)/60, 1000*st_sat_[2].data, alpha=0.8, label="Saturday", lw=0.4)
ax.plot(st_fri_[2].times(reftime=ot_fri)/60, 1000*st_fri_[2].data-0.015, alpha=0.8, label="Friday", lw=0.4)
ax.set_ylabel("Ground displacement (mm)")
ax.set_xlabel("Minutes relative to show start")
ax.legend()
ax.set_title(st_sun_[2].get_id())
plt.savefig("day_compare_disp.png", dpi=300, bbox_inches="tight")
```

# 4. Plot spectrogram of friday night, Z-comp data

```python
winlen = 20
dpi = 300
overlap = 0.9
flim = [0.0, 15]
st_fri.copy().remove_response(inventory=inv, output="ACC")  # In terms of acceleration
for tr in st_fri.select(component="Z"):
    sta_inv = inv.select(station=tr.stats.station)
    nx = int(16. * dpi * 0.8)
    ny = int(9. * dpi * 0.7 * 0.5)
    winlen_ = int(winlen * tr.stats.sampling_rate)
    p, f, t = mlab.specgram(tr.data, NFFT=winlen_,
                            Fs=tr.stats.sampling_rate,
                            noverlap=int(winlen_ * overlap),
                            pad_to=next_pow_2(winlen_) * 4)
    df = 1
    dt = max(1, len(t) // nx)
    p = p[::df, ::dt]
    f = f[::df]
    t = t[::dt]

    t += float(tr.stats.starttime+3600)
    delta = t[1] - t[0]
    t = np.arange(UTCDateTime(t[0]).datetime,
                    UTCDateTime(t[-1] + delta * 0.9).datetime,
                    timedelta(seconds=delta)).astype(datetime)
    bol = np.array((f >= flim[0] * 0.9, f <= flim[1])).all(axis=0)
    vmin = np.percentile(10 * np.log10(p[bol, :]), q=40, axis=None)
    vmax = np.percentile(10 * np.log10(p[bol, :]), q=100, axis=None)

    fig = plt.figure(figsize=(12, 5))
    ax = plt.subplot()
    c= ax.pcolormesh(t, f[bol], 10 * np.log10(p[bol, :]),
                     vmin=vmin, vmax=vmax,
                     shading='nearest',
                     cmap='inferno')
    plt.colorbar(c)
    locator = mdates.AutoDateLocator(minticks=9, maxticks=14)
    formatter = mdates.ConciseDateFormatter(locator)
    ax.xaxis.set_major_locator(locator)
    ax.xaxis.set_major_formatter(formatter)
    ax.set_title(f"{tr.get_id()}")
    ax.set_xlabel("Time (BST)")
    ax.set_ylabel("Beats per second")

    plt.savefig("spec.png", dpi=300, bbox_inches="tight")
```
